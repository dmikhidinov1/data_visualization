## Analysis of Pokemon Type strength

```python
pd.options.display.float_format = '${:,.2f}'.format
#creating a dataframe grouped by Type1 and averaged by the mean of HP

a = df.groupby("Type 1")
mean_df = a.mean()
mean_df = mean_df.reset_index()
```

HP

```python
#now lets create bar chart
fig, ax = plt.subplots(figsize=(15,4)) #assigning the figure size
bars = ax.bar(mean_df['Type 1'], mean_df['HP'], width=0.5) #assigning x and y axis of chart from the mean_df dataframe
#setting names of labes and chart title
ax.set_xlabel('Pokemon Type')
ax.set_ylabel('Average HP')
ax.set_title('Average HP of Pokemons from each Primary Type')
#the following is the code that allowed me to print the labels for each bar
for bar in bars:
    height = bar.get_height()
    label_x_pos = bar.get_x() + bar.get_width() / 2
    ax.text(label_x_pos, round(height,2), s=f'{round(height,2)}', ha='center', va='bottom')
    # above I am also using round function to to show short version of values instead of scientific notation values
```

## Top 10 strongest Dragon Type Pokemons

```python
dragon_df = df[df["Type 1"] == "Dragon"]
dragon_df = dragon_df.nlargest(10, ['BST'])
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(figsize=(8,4))
bars = ax.barh(dragon_df["Name"], dragon_df["BST"], align='center')
ax.set_xlabel('Total of Base Stats')
ax.set_title('Top 10 the strongest Dragon type Pokemons')
for index, value in enumerate(dragon_df["BST"]):
    plt.text(value, index, str(value))
```

```python
#correlation between stats
import seaborn as sns
pokemon_stats = df[(df['Final Evolution']==1)&(df['Legendary']==0)&(df['Mega Evolution']==0)][['HP', 'Att', 'Def', 'Spa', 'Spd', 'Spe']]
g = sns.PairGrid(pokemon_stats)
g.map(sns.regplot,scatter_kws = {'alpha': 0.5,'s':3})
plt.show()
plt.close()
```

```python
dragon_catch = df.loc[df['Type 1'] == "Dragon"]
psychic_catch = df.loc[df['Type 1'] == "Psychic"]
#df[df["Type 1"] == "Psychic"]["Catch Rate"]
steel_catch = df.loc[df['Type 1'] == "Steel"]

fig, ax = plt.subplots(figsize=(15,9))
ax.scatter(dragon_catch["BST"], dragon_catch["Catch Rate"], alpha=.5)
ax.scatter(psychic_catch["BST"], psychic_catch["Catch Rate"], alpha=.5)
ax.scatter(steel_catch["BST"], steel_catch["Catch Rate"], alpha=.5)
ax.set_title('Scatter plot Base stats&Catch Rate', fontsize = 16)
ax.set_xlabel('Base Stats')
ax.set_ylabel('Catch Rate')
```

Next I am looking and 3 strongest Pokemon Types Dragon, Psychic and Steel and their Attack/Defense scatter plot

```python
plt.figure(figsize=(12, 8))

plt.scatter(dragon_catch["Att"], dragon_catch["Def"], c='green', label='Dragon')
plt.scatter(psychic_catch["Att"], psychic_catch["Def"], c='#FF7F7F', label='Psychic')
plt.scatter(steel_catch["Att"], steel_catch["Def"], c='blue', label='Steel')
plt.legend()
plt.title('Scatter plot Attack & Defense for Dragon, Psychic and Steel Pokemons')
plt.xlabel('Attack Stat')
plt.ylabel('Defense')
plt.show()
```

Another interesting analysis I wanted to do was to see what is the catch rate of Pokemon Types that reached the final evolution

```python
final_evolution = df.loc[(df['Final Evolution'] == 1)]

plt.figure(figsize=(15,8), dpi = 80)
sns.boxplot(x = "Type 1", y = "Catch Rate", data=final_evolution, notch=False)

# I found this function online for finding numbers of observations 
def add_n_obs(df,group_col,y):
    medians_dict = {grp[0]:grp[1][y].median() for grp in df.groupby(group_col)}
    xticklabels = [x.get_text() for x in plt.gca().get_xticklabels()]
    n_obs = df.groupby(group_col)[y].size().values
    for (x, xticklabel), n_ob in zip(enumerate(xticklabels), n_obs):
        plt.text(x, medians_dict[xticklabel]*-0.1, ""+str(n_ob), horizontalalignment='center', fontdict={'size':8}, color='black')

add_n_obs(df,group_col='Type 1',y='Catch Rate')    
plt.title('Catch Rate of Pokemon Types that reached Final Evolution', fontsize=22)
plt.show()
```
